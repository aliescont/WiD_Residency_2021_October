```python
# Import libraries for analysis and visualization.
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create dataframe to read in Short Term Occupational Employment Projections for California for 2020-2022
df_STOcc = pd.read_csv("Short-Term_Occupational_Employment_Projections.csv")
# remove spaces in columns name
df_STOcc.columns = df_STOcc.columns.str.replace(' ','_')
df_STOcc.head()
```

```python
# Create new dataframe with total occupations dropped.
df_STOcc2 = df_STOcc.drop([0])

df_STOcc2.head(20)
```

```python
# Reset index and drop index from previous dataframe
df_STOcc2 = df_STOcc2.reset_index(drop=True)
df_STOcc2.head()
```

```python
# Find information about the data file.
df_STOcc2.info()
```

```python
# Remove SOC_Level data to only level 4 to get most detailed occupation titles.
df_STOcc2_level4 = df_STOcc2.loc[df_STOcc2['SOC_Level'] == 4].copy()

df_STOcc2_level4.info()
```

```python
# Summary statistics for numerical columns
df_STOcc2_level4.describe()
```

```python
# Create boxplot of data for visualization of data spread.
df_STOcc2_level4['Numeric_Change'].plot(kind='box', figsize=(15,8))
plt.title('Box Plot of Spread of Data for Numeric Change')
plt.ylabel('Projected Numeric Change')

plt.show()
```

```python
# Sort dataframe by the estimated numeric change
df_STOcc2_level4.sort_values(by='Numeric_Change', ascending=False, inplace=True)
df_STOcc2_level4.head(10)
```

```python
# Create dataframes for 10 occupations with highest and 10 with lowest estimated numeric change
df_top10 = df_STOcc2_level4.head(10)
df_lowest10 = df_STOcc2_level4.tail(10)
df_top10
```

```python
# Create horizontal bar chart of top 10 occupations with highest projected numeric change from 2020 to 2022
df_top10.plot(x='Occupational_Title', y='Numeric_Change', kind='barh')

plt.title('Top 10 Occupations with Estimated Numeric Change from 2020 to 2022')
plt.xlabel('Numeric Change')
plt.ylabel('Occupational Title')
plt.savefig('Top 10 Occupations Short Term')
plt.show()
```

```python
# Create horizontal bar chart of 10 occupations with lowest projected numeric change from 2020 to 2022
plt.barh(df_lowest10.Occupational_Title, df_lowest10.Numeric_Change)
plt.title('10 Occupations with Lowest Estimated Numeric Change from 2020 to 2022')
plt.xlabel('Numeric Change')
plt.ylabel('Occupational Title')
plt.show()
```

```python
# Create dataframe for Occupational Title and Numeric Change
df_Changeper100K = df_STOcc2_level4[['Occupational_Title', 'Numeric_Change']].copy()

# Reset index
df_Changeper100K.reset_index(drop=True)
```

```python
# Calculate the projected numeric change in occupation per 100,000 as Numeric Change Per 100K and add column to dataframe

# Create variable for total numeric change
Total_NumericChange = df_Changeper100K['Numeric_Change'].sum()

Numeric_Change100KCalc = df_Changeper100K['Numeric_Change'] / Total_NumericChange * 100000

# Create column for normalized numerica change per 100K
df_Changeper100K['Numeric_Change_Per_100K'] = Numeric_Change100KCalc

# Sort dataframe by normalized numeric change
df_Changeper100K.sort_values(by='Numeric_Change_Per_100K', ascending=False, inplace=True)

df_Changeper100K.head(10)
```

## Rancho Cordova Data EDA and Analyses from American Community Survey 2019

**Note that top 10 highest projected growth occupations are the same for numeric change and numeric change per 100,000.**

```python
# Create dataframe to read in 5 year American Community Survey data for Rancho Cordova for 2019
df_RC_ACSdata = pd.read_stata('ACS_5Y_RConly_11.30.2021.dta')

df_RC_ACSdata.head()
```

```python
# Get information about the dataframe.
df_RC_ACSdata.info()
```

```python
# Get summary statistics for numerical data.
df_RC_ACSdata.describe()
```

```python
# Get unique values for variable 'live' to determine which data to pull for Rancho Cordova
df_RC_ACSdata.live.unique()
```

```python
#Create dataframe containing only data for Rancho Cordova 
df_RConly_ACSdata = df_RC_ACSdata.loc[df_RC_ACSdata['live'] == 'Rancho Cordova'].copy()

df_RConly_ACSdata.head()
```

```python
# Create dataframe of occupations and their SOC codes
df_RC_OccsSOC = df_RConly_ACSdata[['occ_name', 'socp']].copy()
df_RC_OccsSOC.info()
```

```python
# Get list of race and ethnicity values
df_RConly_ACSdata.race_eth.unique()
```

```python
# Correct spelling error in race and ethnicity column. 'White along' should be 'White alone'
df_RConly_ACSdata['race_eth']=df_RConly_ACSdata['race_eth'].str.replace('White along','White alone')

df_RConly_ACSdata.race_eth.unique()
```

```python
# Create dataframe to sum the weighted count of respondents per occupation
df_weighted_Occ = df_RConly_ACSdata.groupby('occ_name', axis=0).sum()

df_weighted_Occ.head()
```

```python
# Get column list for dataframe
print(list(df_weighted_Occ.columns))
```

```python
# Create variable for total weighted count column to determine per 100K
Total_weighted = df_weighted_Occ['pwgtp'].sum()

# Calculation per 100K
RConly_per100KCalc = df_weighted_Occ['pwgtp'] / Total_weighted * 100000

#Create column for normalized weight per 100K
df_weighted_Occ['Occupations_Per_100K'] = RConly_per100KCalc

# Sort dataframe by normalized weighted count for occupation per 100K
df_weighted_Occ.sort_values(by='Occupations_Per_100K', ascending=False, inplace=True)
#df_weighted_Occ.drop(labels='Occupation_Change_Per_100K', axis=1, inplace=True)

df_weighted_Occ.head(10)
```

```python
print(Total_weighted)
```

```python
# Create a list of the top 10 occupations of Rancho Cordova residents from 2019
df_RCtop10 = df_weighted_Occ.reset_index()
RCtop10_occ = df_RCtop10['occ_name']
RCtop10_occ.head(10)
```

```python
# Create dataframe of Top 10 Highest Growth Occupations per 100K for Rancho Cordova that match Sacramento Projected for 2028
df_RConly_Top10Growth = df_RCtop10[['occ_name', 'Occupations_Per_100K']].copy()

# Reset the index and drop previous index column
df_RConly_Top10Growth.reset_index(drop=True)
df_RConly_Top10Growth.head(10)
```

```python
# Create dataframe of Top 10 Highest Projected Growth Occupations from Rancho Cordova current residents' occupations


RC_Sac_Top10Match = df_RConly_Top10Growth.loc[df_RConly_Top10Growth['occ_name'].isin(['Personal care aides', 'Laborers and freight, stock, and material movers, hand', 'Management analysts', 'Registered nurses', 'Landscaping and groundskeeping workers', 'Cooks', 'Waiters and waitresses', 'General and operations managers', 'Medical assistants', 'Carpenters'])].copy()
RC_Sac_Top10Match.head(10)
```

```python
# Sort occupations names alphabetically

RC_Sac_Top10Match.sort_values(by='occ_name', inplace=True)
RC_Sac_Top10Match.reset_index(drop=True, inplace=True)
RC_Sac_Top10Match.head(10)
```

```python
# List column titles 
print(list(df_RCtop10.columns))
```

```python
# Create dataframe without the mismatched occupations in the projected top10 high growth 2028 dataframe

df_short_LT = df_LTChangeper100K.loc[df_LTChangeper100K['Occupational_Title'].isin(['Personal Care Aides', 'Laborers and Freight, Stock, and Material Movers, Hand', 'Management Analysts', 'Registered Nurses', 'General and Operations Managers', 'Cooks, Restaurant', 'Landscaping and Groundskeeping Workers', 'Waiters and Waitresses', 'Medical Assistants', 'Carpenters'])].copy()
df_short_LT.sort_values(by='Occupational_Title', inplace=True)

df_short_LT.head(10)
```

```python
df_short_LT.reset_index(drop=True, inplace=True)

df_short_LT.head(10)
```

```python
# Concatenate Rancho Cordova data and Projected Sacramento Area data for Highest Growth Occupations
frames = [RC_Sac_Top10Match, df_short_LT]

RC_Sac_Top10Match_Final = pd.concat([RC_Sac_Top10Match, df_short_LT.reindex(RC_Sac_Top10Match.index)], axis=1)
RC_Sac_Top10Match_Final.sort_values(by='Projected_Emp_Estimate_Per_100K', ascending=False, inplace=True)

display(RC_Sac_Top10Match_Final)
```

```python
# Create multiple bar chart comparing number of residents per 100K in occupations with the highest projected growth in the Sacramento County Area
RC_Sac_Top10Match_Final.plot(x='Occupational_Title', y=['Occupations_Per_100K', 'Projected_Emp_Estimate_Per_100K'], kind='barh')

plt.title('Projected High Growth Occupations in 2028 for Sacramento County Area and Current Rancho Cordova Residents')
plt.xlabel('Employment per 100K')
plt.ylabel('Occupational Title')
plt.legend(['Rancho Cordova', 'Projected Sacramento Area'])
plt.show()
```
